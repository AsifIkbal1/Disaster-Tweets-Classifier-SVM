**preproccessing text**

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
```

```python
stop_word=stopwords.words('english')
```

```python
print(stop_word)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
def clean(text):
    stop=' '.join(i for i in text.split(' ') if i not in stop_word)
    lamm=[PorterStemmer().stem(s) for s in stop]
    return stop
```

```python
data_tr['text']=data_tr['text'].apply(clean)
```

```python
data_tr['text']
```

```python
import re
words=[]
for w in range(len(data_tr['text'])):
    sentece=re.sub('[^a-zA-Z]'," ",data_tr['text'][w])
    s=sentece.lower()
    words.append(s)
```

```python
words
```

```python
x=words
y=data_tr['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.10,random_state=44)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
clf=CountVectorizer()
x_train=clf.fit_transform(x_train)
x_test=clf.transform(x_test)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel = 'rbf')
model.fit(x_train,y_train)
```

```python
y_pre=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score ,classification_report,confusion_matrix
```

```python
score=accuracy_score(y_test,y_pre,normalize=True)
```

```python
print(f'Accuracy : {round(score, 4)*100}%')
```

```python
print("classification_report is : \n",classification_report(y_test,y_pre))
```

```python
print("confusion_matrix is : \n" , confusion_matrix(y_test,y_pre))
```
